```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
dataset = pd.read_csv(r"D:\Datasets for ML\creditcard.csv")
dataset.head()
```

```python
# Getting 0 values
legit = dataset[dataset['Class'] == 0]
fraud = dataset[dataset['Class'] == 1]
```

```python
# Calculating the mean values of Class group
dataset.groupby('Class').mean()
```

```python
# Taking random samples
legit_sample = legit.sample(n = 492)
```

```python
new_data = pd.concat([legit_sample,fraud])
new_data.head()
```

```python
new_data['Class'].value_counts()
```

```python
# Now to check whether we got a good sample or not through mean values
new_data.groupby('Class').mean()
```

```python
# Splitting data
X =dataset.iloc[:,:-1]
Y = dataset['Class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42, stratify=Y)
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
print(lr.score(x_train,y_train)*100)
print(lr.score(x_test,y_test)*100)
```

```python
lr.predict(x_test)
```
